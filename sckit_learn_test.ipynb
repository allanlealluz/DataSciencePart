```python
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np 
import seaborn as sns 

df = pd.read_csv("heart-disease.csv")
df.head()
```

```python
df['target'].value_counts(normalize=True).plot(kind="bar", color=["red","blue"])
```

```python
df.info()
```

```python
pd.crosstab(df["target"],df["sex"]).plot(kind="bar", figsize=(10,6), color=["red","blue"])
plt.title("Frequencia cardiaca por sexo")
plt.xlabel("0 - não tem doença, 1 - tem doença")
plt.ylabel("Amostragem")
plt.legend(["Mulher","Homem"])
```
